```python
import yfinance

data = yfinance.Ticker("AAPL").history("6mo")
cierre = data.Close
cierre.plot()
max = round(cierre.max(), 2)
min = round(cierre.min(), 2)
media = round(cierre.mean(), 2)
```

```python
print(max)
print(min)
print(media)
```
